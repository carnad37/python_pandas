# 한국어 말뭉치
KoNLPy에서는 대한민국 헌법 말뭉치인 kolaw와 국회법안 말뭉치인 kobill을 제공  
각 말뭉치가 포함하는 파일의 이름은 fields 메서드로 알 수 있고 open 메서드로 해당 파일의 텍스트를 읽음

## 1. Read document

```python
from konlpy.corpus import kobill    # Docs from pokr.kr/bill
files_ko = kobill.fileids()         # Get file ids
doc_ko = kobill.open('1809890.txt').read()
```

## 2. Tokenize

```python
from konlpy.tag import Okt; t = Okt()
tokens_ko = t.morphs(doc_ko)
tokens_ko
```

## 3. Text 클래스
nltk.Text()는 문서 하나를 편리하게 탐색할 수 있는 다양한 기능을 제공

```python
import nltk
text = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')   # For Python 2, input `name` as u'유니코드'
```

###  lot 메소드를 사용하면 각 단어(토큰)의 사용 빈도를 그래프로 그려준다.

```python
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
font_fname = 'C:\Windows\Fonts\HMKMG.TTF'     # A font of your choice
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)
```

```python
text.plot(20)
plt.show()
```

```python
text.count('초등학교')   # Counts occurrences
```

###  dispersion_plot 메서드는 단어가 사용된 위치를 시각화

```python
text.dispersion_plot(["육아휴직", "개정", "법률", "발의"])
```

###  concordance 메서드로 단어가 사용된 위치를 직접 표시

```python
text.concordance('법률')
```

### collocations
collocations 메서드로는 같이 붙어서 쓰이는 단어 즉, 연어(collocation)를 찾는다.

```python
text.collocations()
```

### similar 메서드는 해당 단어와 비슷한 문맥에서 사용된 단어

```python
text.similar('법률', 10)
```

## 4. FreqDist

FreqDist 클래스는 문서에 사용된 단어(토큰)의 사용빈도 정보를 담는 클래스이다.  
Text 클래스의 vocab 메서드로 추출할 수 있다.  
voca() 빈도 분포를 알수 있다.


```python
fd = text.vocab()
fd
```

 전체 단어의 수, "법률"라는 단어의 출현 횟수, 확률을 각각 계산

```python
from nltk import FreqDist
fd.N(), fd["법률"], fd.freq("법률")
```

most_common 메서드를 사용하면 가장 출현 횟수가 높은 단어를 찾는다.

```python
fd.most_common(5) 
type(fd.most_common(5))
```

items()를 사용하면 빈도분포의 item 전체를 set의 형태로 볼 수 있다.
