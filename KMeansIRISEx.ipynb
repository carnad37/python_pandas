```python
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()

labels = pd.DataFrame(iris.target, columns = ['label'])
data = pd.DataFrame(iris.data, columns = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width'])
data1 = data.join(labels, how='inner')
data1
```

```python
x = 'Sepal_length'
y = 'Sepal_width'
feature = data1[['Sepal_length', 'Sepal_width']]
feature.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = KMeans(n_clusters=3, algorithm='auto')
#algorithm은 auto=full=default이다. 그외의 값으론 elkan이 있다.(자세한건 논문참조)
model.fit(feature)
predict = pd.DataFrame(model.predict(feature), columns = ['predict'])

r = pd.concat([feature, predict], axis = 1)
r
```

```python
centers = pd.DataFrame(model.cluster_centers_, columns = ['Sepal_length', 'Sepal_width'])
#cluster_centers_엔 n_clusters에서 설정한 숫자만큼의 cluseter center의 위치값이 담겨져 있다.

center_x = centers['Sepal_length']
center_y = centers['Sepal_width']

plt.scatter(r[x], r[y], c = r['predict'], alpha = 0.5)
plt.scatter(center_x, center_y, marker='D', c='r')
plt.xlabel(x)
plt.ylabel(y)
plt.show()
```

```python
ct = pd.crosstab(data1['label'], r['predict'])
print(ct)
```

```python
ks = range(1,20)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(feature)
    inertias.append(model.inertia_) #inertia_군집도를 나타내는 값.
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
