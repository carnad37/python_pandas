```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import pandas as pd

font_fname = 'C:/Windows/Fonts/HYGPRM.TTF'
font_family = font_manager.FontProperties(fname=font_fname).get_name()

plt.rcParams["font.family"] = font_family
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 10.
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.labelsize'] = 20.


df = pd.read_csv("iris.csv")
data = df.iloc[:,1:3]
# print(iris_df)
N = 10  # 앞의 10송이만 선택
X = data.iloc[:N, :]  # 꽃받침의 길이와 꽃받침의 폭만 선택

plt.plot(X.T, 'o:')
plt.xticks(range(4), ["꽃받침의 길이", "꽃받침의 폭"])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title("붓꽃의 크기 특성")
plt.legend(["표본 {}".format(i + 1) for i in range(N)])


plt.show()
```

```python

from sklearn.decomposition import PCA

pca1 = PCA(n_components=1)
X_pca  = pca1.fit_transform(X)
print(X_pca)


X_new = pca1.inverse_transform(X_pca)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```
