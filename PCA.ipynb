```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
from sklearn.decomposition import PCA

# from sklearn.datasets import load_iris

font_fname = 'C:/Windows/Fonts/HYKANM.TTF'
font_family = font_manager.FontProperties(fname = font_fname).get_name()

plt.rcParams["font.family"] = font_family
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 10.
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.labelsize'] = 20.
```

```python
df = pd.read_csv("iris.csv")
# load_iris()
N = 20
data = df.iloc[:, 1:3]
X = df.iloc[:N, 1:3]
print(X)
```

```python
plt.plot(X.T, 'o:')
plt.xticks(range(4), ["꽃받침의 길이", "꽃받침의 폭"])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title("붓꽃의 크기 특성")
plt.legend(["표본 {}".format(i + 1) for i in range(N)])
plt.show()
```

```python
plt.scatter(X['SepalLength'], X['SepalWidth']) #x1이 x축, x2가 y축
```

```python
pca1 = PCA(n_components = 'mle')
X_pca = pca1.fit_transform(X)
X_new = pca1.inverse_transform(X_pca)
print(X)
print(X_new)
plt.scatter(X.iloc[:,0], X.iloc[:, 1], alpha = 0.4)
plt.scatter(X_new[:,0], X_new[:, 1], alpha = 0.8)
plt.axis('equal')
```
