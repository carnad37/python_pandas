```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np

x, y = make_blobs(random_state = 1)
kmean = KMeans(n_clusters=3)
kmean.fit(x)
pd.DataFrame(x).head(10)
```

```python
mglearn.discrete_scatter(x[:, 0], x[:, 1], kmean.labels_, markers='o')

mglearn.discrete_scatter(
    kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1], y = np.unique(kmean.labels_),
    markers='^', markeredgewidth=2)
plt.show()
print(kmean.labels_)
print( np.unique(kmean.labels_))
```

```python
fig, axes = plt.subplots(2, 2, gridspec_kw={'hspace' : 0.5})

n_clusters_set = [2, 4, 5, 6]

print(axes.ravel())

for n, ax in zip(n_clusters_set, axes.ravel()):
    kmeans = KMeans(n_clusters=n) # 클러스터 갯수
    kmeans.fit(x)
    mglearn.discrete_scatter(x[:, 0], x[:, 1], y=kmeans.labels_, ax=ax) # x, y, group, ax = plot객체
    mglearn.discrete_scatter(
        kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], y = np.unique(kmeans.labels_),
        markeredgewidth=3, ax = ax,s=10)
    ax.set_title('n_cluster: {}'.format(n))
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')

# for n, ax in zip(n_clusters_set, axes.ravel()):

#     kmeans = KMeans(n_clusters=n) # 클러스터 갯수

#     kmeans.fit(x)

#     mglearn.discrete_scatter(x[:, 0], x[:, 1], y=kmeans.labels_, ax=ax) # x, y, group, ax = plot객체

#     mglearn.discrete_scatter(
# 
#         kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], y=np.unique(kmeans.labels_), # x, y, group

#         markeredgewidth=3, ax=ax, s=15) # 두께, 그림객체, 사이즈

#     ax.set_title('n_cluster: {}'.format(n)) # title

#     ax.set_xlabel('feature 0')

#     ax.set_ylabel('feature 1')



plt.show()
```

```python
x1, y1 = make_blobs(n_samples = 200, cluster_std = [1, 2.5, 0.5], random_state = 170)
#make_blobs는 랜덤값을 생성한다(df의 형태)
#cluster_std는 클러스터의 표준편차. 디폴트는 1이다
#random_state는 난수발생 시드값이다. 난수발생의 시작점과 같고 다음 시드는 새로 생성된 난수 값이 된다.
y_pred = KMeans(n_clusters= 3, random_state=0,).fit_predict(x1)
#n_cluster는 생성하는 클러스터의 갯수이다.
mglearn.discrete_scatter(x1[:, 0], x1[:, 1], y = y_pred)
plt.legend(['cluster 0', 'cluster 1', 'cluster 2'])
plt.xlabel('feature 0')
plt.ylabel('feature 1')
plt
#밀도차로 인한 실패로 간주
```
