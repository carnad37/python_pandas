## 3.다음의 기능을 수행하세요.

```python
import numpy as np
import pandas as pd
docs = np.array([
        'The sun is shining',
        'The weather is sweet',
        'The sun is shining and the weather is sweet'])
```

### 3-1.모두 소문자로 변환

```python
docs_lower = list(map(lambda x: x.lower(), docs))
docs_lower
```

### 3-2.30개 문자(빈칸 포함)만 출력

```python
full_docs = " ".join(docs_lower)
target_word = full_docs[:30]
target_word
```

### 3-3.추출한 Token의 빈도를 출력

#### 3-3-1.30개 문자열 토큰만 포함하는 경우

```python
#vocab()을 통해 구한 FreqDist를 이용한 방법
docs_token = nltk.word_tokenize(" ".join(docs_lower))
docs_text = nltk.Text(docs_token)
docs_vocab = docs_text.vocab()

target_token = nltk.word_tokenize(target_word)
for target in target_token:
    print("%s : %d" % (target, docs_vocab.get(target)) )
```

```python
#CountVectorizer를 이용한 방법
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(tokenizer = nltk.word_tokenize)
tdm = cv.fit_transform(docs_lower)
cv.get_feature_names()
```

```python
target_index = list(map(lambda x: cv.vocabulary_.get(x), target_token))
target_data = list(map(lambda x: tdm.sum(axis=0).flat[x], target_index))
word_count = pd.DataFrame({'단어' : target_token, '빈도' : target_data})
word_count
```

#### 3-3-2.모든 문자열 토큰을 포함하는 경우

```python
word_count = pd.DataFrame(
    {'단어' : cv.get_feature_names(),
     '빈도' : tdm.sum(axis=0).flat
    }
)
word_count.sort_values('빈도', ascending=False)
```

### 3-4.TF-IDF값 출력

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer()
tfidf_tdm = tfidf.fit_transform(tdm)
words = cv.get_feature_names()

count = 1
for unit_tdm in tfidf_tdm:
    print("=== %d.sentence ======================"%count)
    for i, n in sorted(zip(unit_tdm.indices, unit_tdm.data)):
        print("%s : %.5f"%(words[i], n))
    print()
    count += 1
```

## 4.다음의 기능을 수행하세요.

### 4-1.kr-Report_2018.txt 자료를 불러온다

```python
open_text = open('kr-Report_2018.txt', 'r', encoding='utf8')
text = []
while True:
    line = open_text.readline()
    if not line: break
    line = line.strip()
    text.append(line)
open_text.close()
print(text)
# len(text)
```

### 4-2.모두 소문자로 변환

```python
text_lower = list(map(lambda x: x.lower(), text))
text_lower
```

### 4-3.30개 문자(빈칸 포함)만  출력

```python
full_docs = " ".join(text_lower)
target_word = full_docs[:30]
target_word
```

### 4-4.추출한 Token의 빈도를 출력

```python
#CountVectorizer를 이용한 방법
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(tokenizer = nltk.word_tokenize)
tdm = cv.fit_transform(text_lower)
cv.get_feature_names()
word_count = pd.DataFrame(
    {'단어' : cv.get_feature_names(),
     '빈도' : tdm.sum(axis=0).flat
    }
)
word_count.sort_values('빈도', ascending=False)
```

### 4-5.td_idf 값 출력

```python
tfidf = TfidfTransformer()
tfidf_tdm = tfidf.fit_transform(tdm)
words = cv.get_feature_names()

for i, n in sorted(zip(tfidf_tdm[0].indices, tfidf_tdm[0].data)):
    print(words[i], n)
```
