```python
import pandas as pd
import numpy as np
import seaborn as sns

titanic = sns.load_dataset("titanic")
```

```python
df = pd.DataFrame(titanic)
df.head(20)
```

```python
#전처리
#각 컬럼별 NaN의 갯수
df.isnull().sum()
```

```python
#NaN이 3개 이상한 값 제거
res_df = df.dropna(thresh = len(df.columns) - 3)
#fare의 경우 평균값으로 채우기
res_df.fare.fillna(res_df.fare.mean(), inplace=True)
#age는 중앙값으로 채우기
res_df.age.fillna(res_df.age.median(), inplace=True)
#그 외는 최빈값으로 채우기
res_df = res_df.apply(lambda series: series.fillna(series.mode()[0]))
```

```python
res_df.isnull().sum()
```

```python
res_df.head(20)
```

```python
#승선자의 평균 나이
round(res_df.age.mean())
```

```python
#클래스별 승선인원
class_type = list(res_df["class"].unique())
class_type.sort()
class_person = list(map(lambda target: len(res_df[res_df["class"] == target]), class_type))
class_person_series = pd.Series(index = class_type, data = class_person)
class_person_series
```

```python
#클래스별 승선인원(Groupby)
survived_group = res_df.groupby("class")["survived"]
all_passenger_series = survived_group.count()
all_passenger_series
```

```python
#클래스별 생존자의 수
alive_class = list(map(lambda target: len(res_df[(res_df["class"] == target) & (res_df["alive"] == "yes")]), class_type))
alive_class_series = pd.Series(index = class_type, data = alive_class)
alive_class_series
```

```python
#클래스별 생존자 수(Groupby)
survived_series = survived_group.sum()
survived_series
```

```python
#클래스별 생존율
alive_rate_class = alive_class_series.div(class_person_series)
alive_rate_class = alive_rate_class.map(lambda target: "%d%%" % round(target * 100))
alive_rate_class
```

```python
#클래스별 생존율(groupby)
class_survived_rate = round(survived_series.div(all_passenger_series) * 100)
class_survived_rate
```

```python
#생존자중 남성과 여성의 비율
alive_df = res_df[res_df["alive"] == "yes"]
man_alive = len(alive_df[alive_df["sex"]=="male"])
woman_alive = len(alive_df[alive_df["sex"]=="female"])
if man_alive > woman_alive:
    print("여성 : 남성\n1 :", round(man_alive / woman_alive, 2))
else:    
    print("남성 : 여성\n1 :", round(woman_alive / man_alive, 2))
```

```python
#생존자중 남성과 여성의 비율(groupby)
group_by_sex = res_df.groupby("sex")["survived"]
survived_by_sex = group_by_sex.sum()
survived_rate_by_sex = round(survived_by_sex / min(survived_by_sex), 3)
survived_rate_by_sex
```

```python
#생존자의 비율중 남자, 여자, 아이의 비율
child_alive = len(alive_df[alive_df["who"]=="child"])
man_alive = len(alive_df[alive_df["who"]=="woman"])
woman_alive = len(alive_df[alive_df["who"]=="man"])
alive_list = [("여성", woman_alive), ("남성", man_alive), ("아이", child_alive)]
alive_list = sorted(alive_list, key=lambda target: target[1])
second_rate = round(alive_list[1][1] / alive_list[0][1], 2)
third_rate = round(alive_list[2][1] / alive_list[0][1], 2)
print("%s : %s : %s\n 1 : %.2f : %.2f" % (alive_list[0][0], alive_list[1][0], alive_list[2][0], second_rate, third_rate))
```

```python
#생존자의 비율중 남자, 여자, 아이의 비율(groupby)
group_by_who = res_df.groupby("who")["survived"]
survived_by_who = group_by_who.sum()
survived_rate_by_who = round(survived_by_who / min(survived_by_who), 3)
survived_rate_by_who
```
